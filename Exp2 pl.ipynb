```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
df = pd.read_csv("Titanic.csv")
```

```python
print(df.head())
print(df.info())
print(df.describe())
```

```python
print(df.isnull().sum())
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.isnull(), cmap='viridis', cbar=False)

plt.title("Missing Values 24EU01108")
plt.show()
```

```python
sns.histplot(df['Age'], bins=30, kde=True)

plt.title("Age Distribution 24EU01108")
plt.show()
```

```python
sns.histplot(df['Fare'], bins=30, kde=True)

plt.title("Fare Distribution 24EU01108")
plt.show()
```

```python
sns.boxplot(x=df['Age'])

plt.title("Age Boxplot 24EU01108")
plt.show()

sns.boxplot(x=df['Fare'])

plt.title("Fare Boxplot 24EU01108")
plt.show()
```

```python
sns.countplot(x='Survived', data=df)

plt.title("Survival Count 24EU01108")
plt.show()
```

```python
sns.countplot(x='Sex', hue='Survived', data=df)

plt.title("Survival by Gender 24EU01108")
plt.show()
```

```python
plt.figure(figsize=(8,6))

sns.heatmap(df.corr(numeric_only=True),
            annot=True,
            cmap='coolwarm')

plt.title("Correlation Matrix 24EU01108")
plt.show()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())

df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

df['Fare'] = df['Fare'].fillna(df['Fare'].median())
```

```python
df.drop('Cabin', axis=1, inplace=True)
```

```python
df.drop(['PassengerId','Name','Ticket'],
        axis=1,
        inplace=True)
```

```python
le = LabelEncoder()

df['Sex'] = le.fit_transform(df['Sex'])

df['Embarked'] = le.fit_transform(df['Embarked'])
```

```python
scaler = StandardScaler()

df[['Age','Fare']] = scaler.fit_transform(df[['Age','Fare']])
```

```python
print(df.duplicated().sum())

df.drop_duplicates(inplace=True)
```
